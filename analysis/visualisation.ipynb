```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly. express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot()
```

```python
penguins['bill_depth_mm'].plot(linestyle = '',
                               marker='o',
                               markersize = 7,
                               alpha = 0.5,
                               figsize = (7,7),
                               markerfacecolor = 'red',
                               markeredgecolor = 'black')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='bar',figsize=(15,4),color='green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind='area',figsize=(15,4),color='pink')
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind='scatter',x='bill_length_mm', y = 'bill_depth_mm',s = 100,alpha = .5,color='pink')
```

```python
penguins.plot(kind='scatter',x='bill_length_mm', y = 'flipper_length_mm',s = 100,alpha = .5,color = 'green')
```

```python
fig,axs = plt.subplots(1,2,figsize=(15,4),constrained_layout = True)
penguins.plot(kind='scatter',x='bill_length_mm', y = 'bill_depth_mm',s = 100,alpha = .5,ax=axs[0],title='penguins bill data',color='pink')
penguins.plot(kind='scatter',x='bill_length_mm', y = 'flipper_length_mm',s = 100,alpha = .5,color = 'green',ax=axs[1],title = 'penguin flipper')
plt.show()
```
